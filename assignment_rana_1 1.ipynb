```
#(25 points). Write a program that will do the following string formatting tasks:

#(1) Ask the user to enter two sentences.

#(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

#(3) Compare the two sentences. Print out a list of words that occur in both sentences.

#(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

#(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc
```

```
import string
pun = set(string.punctuation)

#question1
fir_input = str(input(" 1st sentences : "))
fir = fir_input.split(' ')
second_input = str(input("2nd sentences : "))
second = second_input.split(' ')

print('1st sentence splitted into string ')
print(fir)
print('2nd sentence splited into string')
print(second)

print('Common Elements in lists:')
common_elements = list(set(fir).intersection(second))
print(common_elements)

print('1st Sentence without punctuation')
fir_words_stripped = ''.join(x for x in fir_input if x not in pun)
print(fir_words_stripped)
print('2ndSentence without punctuation')
second_words_stripped = ''.join(x for x in second_input if x not in pun)
print(second_words_stripped)

#question5
def Check_Vow(string, vowels): 
    count = {}.fromkeys(vowels, 0) 
    for character in string: 
        if character in count: 
            count[character] += 1    
    return count 
       
vowels = 'aeiou'
fir_input_vowels = Check_Vow(fir_input, vowels)
second_input_vowels = Check_Vow(second_input, vowels)

print('Fir Sentence vowels')
print (Check_Vow(fir_input, vowels)) 

print('Second Sentence vowels')
print (Check_Vow(second_input, vowels))

print('Both Sentence vowels')
result = {key: fir_input_vowels.get(key, 0) + second_input_vowels.get(key, 0)
          for key in set(fir_input_vowels) | set(second_input_vowels)}
print(result)

```

```
pd.options.mode.chained_assignment = None 
occured_dataframe = dataframe[dataframe['title_field']=='Occupied Housing Units']
occured_dataframe = occured_dataframe.reset_index(drop=False)
occured_dataframe['dif'] = occured_dataframe['value'].diff()
occured_dataframe['combine'] = occured_dataframe.year.astype(str) + '-' +  occured_dataframe.year.astype(str).shift(-1)
occured_dataframe_range = occured_dataframe[['dif','combine']] 
occured_dataframe_range['combine'] =  occured_dataframe_range['combine'].shift(1)
occured_dataframe_range = occured_dataframe_range.dropna()
occured_dataframe_range = occured_dataframe_range.reindex(index=occured_dataframe_range.index[::-1])
occured_dataframe_range = occured_dataframe_range.reset_index(drop=True)
occured_dataframe_range = occured_dataframe_range[['combine','dif']]
occured_dataframe_range.columns = ['year','Diff in Occupied Housing Units ']
occured_dataframe_range
```
