```python
# from six.moves import urllib
# import pandas as pd

# text_array = []
# url = "https://raw.githubusercontent.com/unt-iialab/INFO5731_FALL2020/master/In_class_exercise/01-05-1%20%20Adams%20v%20Tanner.txt"
# text_file = urllib.request.urlopen(url)

# for line in text_file:
#   decoded_data = line.decode("utf-8").replace('\r\n', '')
#   if decoded_data: text_array.append(decoded_data)
# df = pd.DataFrame (text_array, columns = ['Raw Text'])
# print(df)

import pandas as pd
df = pd.read_csv(r'C:\Users\Samhith\Downloads\Bhavya_INFO5731-master\Bhavya_INFO5731-master\Updated\Criminal Law Cases.csv',error_bad_lines=False)
# summary_list = pd.DataFrame(df['Summary'],columns = ['Raw Text'])
summary = pd.DataFrame(df['Summary'])
summary = summary.dropna()
summary = summary.reset_index(drop=True)
summary.isna()
```

**Number of sentences**

```python
summary['Number of Sentences'] = summary['Summary'].apply(lambda x: len(str(x).split(".")))
summary
```

**Number of words**

```python
summary['Number of Words'] = summary['Summary'].apply(lambda x: len(str(x).split(" ")))
summary
```

**Number of characters**

```python
summary['Number of Characters'] = summary['Summary'].str.len()
summary
```

**Average word length**

```python
def avg_word_length(sentence):
  words = sentence.split()
  if len(words) != 0:
    return(sum(len(word) for word in words)/len(words))
  else:
    return None

summary['Average Word Length'] = summary['Summary'].apply(lambda x: avg_word_length(x))
summary
```

**Number of stopwords**

```python
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
```

```python
stop = stopwords.words('english')

summary['Number of Stopwords'] = summary['Summary'].apply(lambda x: len([x for x in x.split() if x in stop]))
summary
```

**Number of special characters**

```python
def special_characters(sentence):
  special = 0
  for value in range(len(sentence)):
    if not((sentence[value].isalpha()) and (sentence[value].isdigit())):
        special = special + 1
  return special

summary['Number of special characters'] = summary['Summary'].apply(lambda x: special_characters(x))
summary
```

**Number of numerics**

```python
summary['Numbers of numerics'] = summary['Summary'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
summary
```

**Number of uppercase words**

```python
summary['Number of upper case words'] = summary['Summary'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
summary
```

## **1.2 Basic Text Pre-processing of text data**

**Lower casing**

```python
summary['Lower Case'] = summary['Summary'].apply(lambda x: " ".join(x.lower() for x in x.split()))
summary
```

**Punctuation removal**

```python
summary['Removal of Punctuation'] = summary['Lower Case'].str.replace('[^\w\s]','')
summary
```

**Stopwords removal**

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
summary['Stopwords Removal'] = summary['Removal of Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
summary
```

**Frequent words removal**

```python
freq_words = pd.Series(' '.join(summary['Stopwords Removal']).split()).value_counts()[:10]
print(freq_words)
freq_words_list = list(freq_words.index)
summary['After common words removal'] = summary['Stopwords Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
summary
```

**Rare words removal**

```python
rare_words = pd.Series(' '.join(summary['Stopwords Removal']).split()).value_counts()[-10:]
print(rare_words)
rare_words_list = list(rare_words.index)
summary['After Rare Words removal'] = summary['After common words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words_list))
summary
```

**Spelling correction**

```python
from textblob import TextBlob
summary['Spelling Correction'] = summary['After Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
summary
```

**Tokenization**

```python
import nltk
nltk.download('punkt')
summary['Tokenization'] = summary['Spelling Correction'].apply(lambda x: TextBlob(x).words)
summary
```

**Stemming**

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
summary['After Stemming'] = summary['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
summary
```

**Lemmatization**

```python
from textblob import Word
import nltk
nltk.download('wordnet')

summary['After Lemmatization'] = summary['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
summary
```

## **1.3 Saving to CSV**

```python
summary.to_csv('cleaned_data.csv',index=False)
```

## **1.4 Advance Text Processing**

**Calculating Term frequencies**

```python
tf1 = (summary['After Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

**Top 10 - one grams**

```python
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

words_sentence = []
for sentence in summary['After Lemmatization']:
  words_sentence.append(word_tokenize(sentence))
aftercleaning = [x for x in words_sentence if x != []]
iterations = list(itertools.chain.from_iterable(aftercleaning))
one_grams = ngrams(iterations, 1)
collections.Counter(one_grams).most_common(10)
```

**Top 10 - Bi grams**

```python
bi_grams = nltk.bigrams(iterations)
frequency_dist = nltk.FreqDist(bi_grams)
(frequency_dist.most_common(10))
```

**Top 10 - Tri grams**

```python
tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
frequency_dist.most_common(10)
```
